Calculate the Pearson's correlation between two variables

```python
## numpy - library for numerical operations like calculating mean, std, covariance, generating random numbers. Also for working with arrays
# importing functions from library
from numpy import mean
from numpy import std
from numpy import cov
from numpy.random import randn
from numpy.random import seed

# matplotlib - for creating plots
from matplotlib import pyplot as plt

#import seaborn as sns

# scipy.stats - for calculating Pearson's correlation coefficient
from scipy.stats import pearsonr

# seed random number generator - sets seed for random number generation - so random numbers generated are reproducible
# seed(1) - every time the code is run, the same random numbers will be generated
seed(1)
```

Prepare data (generate random, correlated data)

```python
# creates an array of 1000 random numbers from a standard normal distribution (mean = 0, std = 1)
# scaled by 20 and shifted by 100
data1 = 20 * randn(1000) + 100

# starting values of data 1 (i.e. always influenced by data1)
# scaled by 10 >> changing the distribution to have a std of 10 (mean remains 0)
# shifts the entire distribution up by 50 (new mean is 50)
# data2 retains the general shape and trend of data1 while introducing variability with the noise
# therefore, correlation between the two datapoints
data2 = data1 + (10 * randn(1000) + 50)
```

Calculate covariance matrix

```python
# computes how much two variables change together
# result = 2x2 matrix where the off-diagonal elements represent the covariance b/w data1 & data2
covariance = cov(data1, data2)
covariance
```

Calculate Pearson's correlation

```python
# measures the linear correlation b/w data1 & data2 (results in a value b/w -1: strong negative correlation and 1:strong positive correlation. 0 = no correlation)
# _ = used to ignore the second returned value (p-value herein)
corr, _ = pearsonr(data1, data2)
corr
```

Plot

```python
# plots (data1 = x axis; data2 = y axis)
plt.scatter(data1, data2)

# displays the plot
plt.show()
```

Summarize key statistics

```python
# %.3f = used in Python string formatting to control how a floating-point number is displayed.
# % = format specifier
# .3 = precision of the floating-point number (display 3 digits after decimal point)
# f = floating point number
print('data1: mean=%.3f std=%.3f' % (mean(data1), std(data1)))
print('data2: mean=%.3f std=%.3f' % (mean(data2), std(data2)))
print('Covariance: %.3f' % covariance[0][1])
print('Pearsons correlation: %.3f' % corr)
```
